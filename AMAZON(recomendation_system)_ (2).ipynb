# train_test_split

```python
from sklearn.model_selection import train_test_split

x= df_final.drop("rating",axis=1,)
y= df_final["rating"]


x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=1)
```

# POPULARITY RECOMENDER MODEL BULINDING

```python
#grouping of train and test data:taking number of user as recomendation score

x_train_grouped= x_train.groupby("product_id").agg({"user_id":"count"}).reset_index()
x_train_grouped.rename(columns={"user_id":"recomendation_score"},inplace=True)

x_train_grouped.head()
```

```python
x_test_grouped= x_test.groupby("product_id").agg({"user_id":"count"}).reset_index()
x_test_grouped.rename(columns={"user_id":"recomendation_score"},inplace=True)

x_test_grouped.head()
```

```python
# the recomendation score

popular_recomendation= x_train_grouped[["recomendation_score","product_id"]].sort_values(by=['recomendation_score',"product_id"],ascending=False)
popular_recomendation["rank"]=popular_recomendation["recomendation_score"].rank(ascending=True)
popular_recomendation.head()
```

```python
popular_test_recomendation = x_test_grouped[["recomendation_score","product_id"]].sort_values(by=['recomendation_score',"product_id"],ascending=False)
popular_test_recomendation ["rank"]=popular_test_recomendation ["recomendation_score"].rank(ascending=True)
popular_test_recomendation .head()
```

```python
def recommend(user_id):
    user_recomendation= popular_recomendation

    user_recomendation['user_id'] = user_id

    colms = user_recomendation.columns.tolist()
    colms = colms[-1:] + colms[:-1]
    user_recomendation = user_recomendation[colms]

    return user_recomendation
```

```python
particular_user_recomendation=[123,125,230]

for i in particular_user_recomendation:
  print(f"the recomenadtion for user{i} is")
  print(recommend(i))
```

# CASE STUDY


1. Introduction

In today's digital marketplace, recommendation systems are essential tools that drive customer engagement and sales. Amazon, being one of the largest online retailers, heavily relies on recommendation engines to suggest products to users based on their preferences and behaviors.
In this project, we built a basic recommendation system using user review data from Amazon, focusing on making personalized product suggestions.

2. Objective
The primary goal of this project was to:
Analyze user ratings for different products.
Build a recommendation system that predicts what products a user may like.


3. Dataset Overview
Source: Amazon product ratings dataset.

Features:

userId: Unique ID for each user.

productId: Unique ID for each product.

ratings: Rating given by the user to the product (usually from 1 to 5).


4. Data Preprocessing
Checked for missing values and found none.
Removed duplicate entries to avoid bias.
Data exploration steps included:
Distribution of ratings.

Top users with the most reviews.

Most popular products based on ratings.

5. Recommendation System Techniques Used
A. Popularity-Based Recommendation
Products are recommended based on overall popularity.

Focuses on products with the highest average ratings and highest number of ratings.

Limitation: Same recommendations for all users (not personalized).

B. Collaborative Filtering-Based Recommendation
Focused on user-product interactions.

Built a User-Item Matrix where rows are users, columns are products, and the values are ratings.

Used a K-Nearest Neighbors (KNN) algorithm:

Find users with similar rating patterns.

Recommend products liked by similar users.



6. Model Evaluation
Since it is a basic model, offline evaluation was done manually by checking:

Relevance of recommended products.


7. Key Observations
Users with high activity (many reviews) dominate the recommendation outputs.



8. Drawback:

Popularity models are easy but lack personalization, while collaborative filtering offers better personalization but needs more computational resources.



9. Conclusion
The recommendation system developed demonstrates the core concepts behind non-personalized suggestions.
While simple methods like popularity-based models provide a quick solution, enhance user exprence;
Building a more scalable, accurate, and robust recommendation system would require integrating more advanced techniques and richer datasets.
